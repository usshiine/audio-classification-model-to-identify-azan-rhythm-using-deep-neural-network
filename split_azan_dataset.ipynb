```python
import os
import shutil
from sklearn.model_selection import train_test_split
import random

# Daftar direktori dataset
dataset_directories = [
    r'C:\Users\ushine\Data Mining\Azan Sukses\combined_dataset\ajam',
    r'C:\Users\ushine\Data Mining\Azan Sukses\combined_dataset\bayati',
    r'C:\Users\ushine\Data Mining\Azan Sukses\combined_dataset\hijaz',
    r'C:\Users\ushine\Data Mining\Azan Sukses\combined_dataset\nahawand',
    r'C:\Users\ushine\Data Mining\Azan Sukses\combined_dataset\rast'
]

# Fungsi untuk membagi dataset dan menyimpannya
def split_and_save_dataset(input_directory, output_train_directory, output_test_directory, test_size=0.2, random_state=random.randint(1, 1000)):
    # Membaca daftar file dalam direktori
    file_list = os.listdir(input_directory)

    # Membagi dataset menjadi training set dan testing set
    train_files, test_files = train_test_split(file_list, test_size=test_size, random_state=random_state)

    # Membuat direktori training jika belum ada
    if not os.path.exists(output_train_directory):
        os.makedirs(output_train_directory)

    # Membuat direktori testing jika belum ada
    if not os.path.exists(output_test_directory):
        os.makedirs(output_test_directory)

    # Memindahkan file ke direktori training
    for file_name in train_files:
        source_path = os.path.join(input_directory, file_name)
        destination_path = os.path.join(output_train_directory, file_name)
        shutil.copyfile(source_path, destination_path)

    # Memindahkan file ke direktori testing
    for file_name in test_files:
        source_path = os.path.join(input_directory, file_name)
        destination_path = os.path.join(output_test_directory, file_name)
        shutil.copyfile(source_path, destination_path)

    print(f"Dataset di {input_directory} berhasil dibagi dan disimpan di:")
    print(f"Training set: {output_train_directory}")
    print(f"Testing set: {output_test_directory}")

# Melakukan pembagian dataset dan penyimpanan untuk setiap direktori dataset
for dataset_directory in dataset_directories:
    output_train_directory = os.path.join(r'C:\Users\ushine\Data Mining\Azan Sukses\split_azan_dataset\training', os.path.basename(dataset_directory))
    output_test_directory = os.path.join(r'C:\Users\ushine\Data Mining\Azan Sukses\split_azan_dataset\testing', os.path.basename(dataset_directory))
    
    split_and_save_dataset(dataset_directory, output_train_directory, output_test_directory)
```
