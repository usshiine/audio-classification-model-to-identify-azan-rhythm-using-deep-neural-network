**Volume Perturbation**

```python
from pydub import AudioSegment
import os

# Fungsi untuk melakukan volume perturbation
def perturb_volume(audio, db_change):
    # Ubah nilai dB
    return audio + db_change

# Fungsi untuk memproses semua file audio dalam folder
def process_folder(input_folder, output_base_folder, db_change):
    # Dapatkan nama subfolder dari input folder
    subfolder_name = os.path.basename(input_folder)

    # Tentukan output folder untuk subfolder yang sama dengan input
    output_folder = os.path.join(output_base_folder, subfolder_name)

    # Pastikan output_folder ada atau buat jika belum ada
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Loop melalui setiap file audio di folder input
    for filename in os.listdir(input_folder):
        if filename.endswith(".wav"):
            # Baca file audio
            audio_path = os.path.join(input_folder, filename)
            original_audio = AudioSegment.from_wav(audio_path)

            # Perturbasi dengan nilai dB yang diberikan
            perturbed_audio = perturb_volume(original_audio, db_change)

            # Ambil nilai perturbasi (untuk menyertakan dalam nama file)
            perturbation_value = f"{db_change:+d}"

            # Simpan audio yang telah diperturbasi ke output_folder dalam format WAV
            output_filename = f"perturbed_{perturbation_value}dB_{filename[:-4]}.wav"
            output_path = os.path.join(output_folder, output_filename)
            perturbed_audio.export(output_path, format="wav")

# Folder input dan output
input_folders = [
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_full_stereo\ajam",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_full_stereo\bayati",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_full_stereo\hijaz",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_full_stereo\nahawand",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_full_stereo\rast",
]

output_base_folder = r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_volume_perturbation"

# Loop melalui setiap folder dan proses file audio
for input_folder in input_folders:
    process_folder(input_folder, output_base_folder, -6)  # Perturbation with -6 dB
    process_folder(input_folder, output_base_folder, 8)   # Perturbation with 8 dB
```

**3 Folders Combined**

```python
import os
import shutil

# Daftar path untuk dataset_speed_pertubation
speed_pertubation_paths = [
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_speed_perturbation\ajam",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_speed_perturbation\bayati",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_speed_perturbation\hijaz",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_speed_perturbation\nahawand",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_speed_perturbation\rast",
]

# Daftar path untuk dataset_gaussian_snr
volume_pertubation_paths = [
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_volume_perturbation\ajam",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_volume_perturbation\bayati",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_volume_perturbation\hijaz",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_volume_perturbation\nahawand",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_volume_perturbation\rast",
]

# Daftar path untuk dataset_speed_gaussian
gaussian_snr_paths = [
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_gaussian_snr\ajam",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_gaussian_snr\bayati",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_gaussian_snr\hijaz",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_gaussian_snr\nahawand",
    r"C:\Users\ushine\Data Mining\Azan Sukses\dataset_gaussian_snr\rast",
]

# Folder untuk menyimpan hasil gabungan
combined_folder = r"C:\Users\ushine\Data Mining\Azan Sukses\combined_dataset"

# Membuat folder untuk menyimpan hasil gabungan
os.makedirs(combined_folder, exist_ok=True)

# Fungsi untuk menggabungkan seluruh file dari dua set path
def combine_files(source_paths, destination_folder):
    for source_path in source_paths:
        category = os.path.basename(source_path)
        destination_path = os.path.join(destination_folder, category)

        # Membuat folder di dalam destination_folder jika belum ada
        os.makedirs(destination_path, exist_ok=True)

        # Menggabungkan seluruh file dari source_path ke destination_path
        for root, dirs, files in os.walk(source_path):
            for file in files:
                source_file = os.path.join(root, file)
                destination_file = os.path.join(destination_path, file)
                shutil.copy2(source_file, destination_file)

# Menggabungkan seluruh file dari kedua set path
combine_files(speed_pertubation_paths, combined_folder)
combine_files(volume_pertubation_paths, combined_folder)
combine_files(gaussian_snr_paths, combined_folder)

print("Gabungan seluruh file berhasil dibuat.")
```
